# Case study: Decision trees

This use case is intended to solve a malware detection problem on Android devices by analysing the network traffic generated by the device using decision trees.

## Dataset: Android malware detection

### Description

The sophisticated and advanced Android malware is able to detect the presence of the emulator used by the malware analyst and, in response, alter its behavior to evade detection. To overcome this issue, we installed the Android applications on a real device and captured its network traffic. See our publicly available Android Sandbox.

The CICAAGM dataset is captured by installing the Android apps on real smartphones in a semi-automated manner. The dataset is generated from 1,900 applications in the following three categories:

**1. Adware (250 apps)**
* Airpush: Designed to deliver unsolicited advertisements to the user’s system for information stealing.
* Dowgin: Designed as an advertisement library that can also steal the user’s information.
* Kemoge: Designed to take over a user’s Android device. This adware is a hybrid of botnet and disguises itself as popular apps via repackaging.
* Mobidash: Designed to display ads and compromise the user’s personal information.
* Shuanet: Similar to Kemoge, Shuanet is also designed to take over a user’s device.

**2. General Malware (150 apps)**
* AVpass: Designed to be distributed disguised as a Clock app.
* FakeAV: Designed as a scam that tricks users into purchasing a full version of the software to remediate non-existing infections.
* FakeFlash/FakePlayer: Designed as a fake Flash app to direct users to a website (after successful installation).
* GGtracker: Designed for SMS fraud (sends SMS messages to a premium-rate number) and information stealing.
* Penetho: Designed as a fake service (a hack tool for Android devices that can be used to crack WiFi passwords). This malware can also infect the user’s computer via infected email attachments, fake updates, external media, and infected documents.

**3. Benign (1500 apps)**
* 2015 GooglePlay market (top free popular and top free new)
* 2016 GooglePlay market (top free popular and top free new)

### Data Files
* pcap files – the network traffic of both malware and benign apps (20% malware and 80% benign)
* .csv files - the list of extracted network traffic features generated by the CIC-flowmeter

## Auxiliary functions

```python
# Construction of a function that performs full partitioning
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
def remove_labels(df, label_name):
    X = df.drop(label_name, axis=1)
    y = df[label_name].copy()
    return (X, y)
```

```python
def evaluate_result(y_pred, y, y_prep_pred, y_prep, metric):
    print(metric.__name__, "WITHOUT preparation:", metric(y_pred, y, average='weighted'))
    print(metric.__name__, "WITH preparation:", metric(y_prep_pred, y_prep, average='weighted'))
```

## 3. Division of the data set

```python
# We split the dataset
train_set, val_set, test_set = train_val_test_split(X)
```

```python
X_train, y_train = remove_labels(train_set, 'calss')
X_val, y_val = remove_labels(val_set, 'calss')
X_test, y_test = remove_labels(test_set, 'calss')
```

## 4. Scaling the dataset

It is important to understand that decision trees are algorithms that **do not require too much data preparation**, namely, they do not require scaling or normalisation. In this exercise we will scale the dataset and compare the results with the unscaled dataset. In this way it is demonstrated how applying preprocessing such as scaling can even affect the performance of the model.

```python
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
scaler = RobustScaler()
X_test_scaled = scaler.fit_transform(X_test)
```

```python
scaler = RobustScaler()
X_val_scaled = scaler.fit_transform(X_val)
```

```python
# Transformation to a Pandas DataFrame
X_train_scaled = DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_train_scaled.head(10)
```

```python
X_train_scaled.describe()
```

## 5. Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

MAX_DEPTH = 10

# Model trained on the unscaled dataset
clf_tree = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=42)
clf_tree.fit(X_train, y_train)
```

```python
# Model trained on the scaled dataset
clf_tree_scaled = DecisionTreeClassifier(max_depth=MAX_DEPTH, random_state=42)
clf_tree_scaled.fit(X_train_scaled, y_train)
```

It is often interesting to start by predicting with the train set itself, in order to check whether overfitting is occurring.

```python
# Predecimos con el conjunto de datos de entrenamiento
y_train_pred = clf_tree.predict(X_train)
y_train_prep_pred = clf_tree_scaled.predict(X_train_scaled)
```

```python
# We compare results between scaled and unscaled
evaluate_result(y_train_pred, y_train, y_train_prep_pred, y_train, f1_score)
```

```python
# We predict with the validation dataset
y_pred = clf_tree.predict(X_val)
y_prep_pred = clf_tree_scaled.predict(X_val_scaled)
```

```python
# We compare results between scaled and unscaled
evaluate_result(y_pred, y_val, y_prep_pred, y_val, f1_score)
```

**At this point, based on the results we have seen, we are left with the option of not scaling our datasets**.

## 6. Visualising the decision limit

```python
# We reduce the number of attributes in the dataset to better visualise it.
X_train_reduced = X_train[['min_flowpktl', 'flow_fin']]
```

```python
# We generate a model with the reduced dataset.
clf_tree_reduced = DecisionTreeClassifier(max_depth=5, random_state=42)
clf_tree_reduced.fit(X_train_reduced, y_train)
```

```python
# We graphically represent the constructed decision boundary
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
%matplotlib inline

def plot_decision_boundary(clf, X, y, plot_training=True, resolution=1000):
    mins = X.min(axis=0) - 1
    maxs = X.max(axis=0) + 1
    x1, x2 = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
    plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="normal")
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="adware")
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="malware")
        plt.axis([mins[0], maxs[0], mins[1], maxs[1]])
    plt.xlabel('min_flowpktl', fontsize=14)
    plt.ylabel('flow_fin', fontsize=14, rotation=90)

plt.figure(figsize=(12, 6))
plot_decision_boundary(clf_tree_reduced, X_train_reduced.values, y_train)
plt.show()
```

```python
# We paint the tree to compare it with the graphical representation above.
from graphviz import Source
from sklearn.tree import export_graphviz
import os

export_graphviz(
        clf_tree_reduced,
        out_file="android_malware.dot",
        feature_names=X_train_reduced.columns,
        class_names=["benign", "adware", "malware"],
        rounded=True,
        filled=True
    )

Source.from_file("android_malware.dot")
```
