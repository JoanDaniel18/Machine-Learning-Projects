## Data set

### Description
NSL-KDD is a data set suggested to solve some of the inherent problems of the KDD'99 data set which are mentioned in. Although, this new version of the KDD data set still suffers from some of the problems discussed by McHugh and may not be a perfect representative of existing real networks, because of the lack of public data sets for network-based IDSs, we believe it still can be applied as an effective benchmark data set to help researchers compare different intrusion detection methods. Furthermore, the number of records in the NSL-KDD train and test sets are reasonable. This advantage makes it affordable to run the experiments on the complete set without the need to randomly select a small portion. Consequently, evaluation results of different research work will be consistent and comparable.



### Data files
* <span style="color:green">**KDDTrain+.ARFF**: The full NSL-KDD train set with binary labels in ARFF format</span>
* KDDTrain+.TXT: The full NSL-KDD train set including attack-type labels and difficulty level in CSV format
* KDDTrain+_20Percent.ARFF:	A 20% subset of the KDDTrain+.arff file
* KDDTrain+_20Percent.TXT:	A 20% subset of the KDDTrain+.txt file
* KDDTest+.ARFF:	The full NSL-KDD test set with binary labels in ARFF format
* KDDTest+.TXT:	The full NSL-KDD test set including attack-type labels and difficulty level in CSV format
* KDDTest-21.ARFF:	A subset of the KDDTest+.arff file which does not include records with difficulty level of 21 out of 21
* KDDTest-21.TXT:	A subset of the KDDTest+.txt file which does not include records with difficulty level of 21 out of 21

## Funciones auxiliares

```python
def load_kdd_dataset(data_path):
    """Reading the NSL-KDD dataset."""
    with open(data_path, 'r') as train_set:
        dataset = arff.load(train_set)
    attributes = [attr[0] for attr in dataset["attributes"]]
    return pd.DataFrame(dataset["data"], columns=attributes)
```

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

## 3 Cleaning the data

Before we start, we are going to retrieve the clean dataset and separate the labels from the rest of the data, we do not necessarily want to apply the same transformations on both sets.

```python
# We separate the input characteristics from the output characteristic.
X_train = train_set.drop("class", axis=1)
y_train = train_set["class"].copy()
```

```python
# To illustrate this section let's add some null values
# to some characteristics of the dataset
X_train.loc[(X_train["src_bytes"]>400) & (X_train["src_bytes"]<800), "src_bytes"] = np.nan
X_train.loc[(X_train["dst_bytes"]>500) & (X_train["dst_bytes"]<2000), "dst_bytes"] = np.nan
X_train
```

Most Machine Learning algorithms cannot work on features containing null values. Therefore, there are three options to replace them: * Delete the corresponding rows * Delete the corresponding attribute (column) * Fill them with a given value (zero, average...)

```python
# We check if there is any attribute with null values
X_train.isna().any()
```

```python
# Select rows containing null values
null_rows_values  = X_train[X_train.isnull().any(axis=1)]
null_rows_values
```

### Option 1: Delete rows with null values

```python
# We copy the dataset so as not to alter the original.
X_train_copy = X_train.copy()
```

```python
# Delete rows with null values
X_train_copy.dropna(subset=["src_bytes", "dst_bytes"], inplace=True)
X_train_copy
```

```python
# We count the number of rows deleted
print("The number of rows deleted is:", len(X_train) - len(X_train_copy))
```

### Option 2: Remove attributes with null values

```python
# We copy the dataset so as not to alter the original.
X_train_copy = X_train.copy()
```

```python
# Remove attributes with null values
X_train_copy.drop(["src_bytes", "dst_bytes"], axis=1, inplace=True)
X_train_copy
```

```python
# We count the number of attributes removed
print("The number of attributes removed is:", len(list(X_train)) - len(list(X_train_copy)))
```

### Option 3: Fill null values with a given value

```python
# We copy the dataset so as not to alter the original.
X_train_copy = X_train.copy()
```

```python
# We fill in the null values with the average of the attribute values
media_srcbytes = X_train_copy["src_bytes"].mean()
media_dstbytes = X_train_copy["dst_bytes"].mean()

X_train_copy["src_bytes"].fillna(media_srcbytes, inplace=True)
X_train_copy["dst_bytes"].fillna(media_dstbytes, inplace=True)

X_train_copy
```

```python
# We copy the dataset so as not to alter the original.
X_train_copy = X_train.copy()
```

```python
# A very high value in the attribute can trigger the average
# We fill in the values with the median
mediana_srcbytes = X_train_copy["src_bytes"].median()
mediana_dstbytes = X_train_copy["dst_bytes"].median()

X_train_copy["src_bytes"].fillna(mediana_srcbytes, inplace=True)
X_train_copy["dst_bytes"].fillna(mediana_dstbytes, inplace=True)

X_train_copy
```

## APIs de sklearn

Before we continue, let's make a brief review of how sklearn APIs work:

**Estimators**: Any object that can estimate some parameter:
* The estimator itself is formed by the fit() method, which always takes a dataset as an argument
* Any other parameter of this method, is a hyperparameter  

**Transformers**: They are estimators capable of transforming the dataset (like Inputer)
* The transformation is performed by the transform() method
* They receive a dataset as input parameter

**Predictors**: They are estimators capable of making predictions
* The prediction is performed by the predict() method
* They receive a dataset as input
* They return a dataset with the predictions
* They have a score() method to evaluate the prediction result

Machine Learning algorithms generally ingest numerical data. In our dataset we have a lot of categorical values and therefore we have to convert them to numerical values.

```python
X_train.info()
```

There are different ways to convert categorical attributes into numerical ones. Probably the simplest is provided by Pandas' **factorize** method. It transforms each category into a sequential number.

```python
protocol_type = X_train['protocol_type']
protocol_type_encoded, categorias = protocol_type.factorize()
```

```python
# We show on the screen how they have been coded
for i in range(10):
    print(protocol_type.iloc[i], "=", protocol_type_encoded[i])
```

```python
print(categorias)
```

### Advanced sklearn transformations

#### Ordinal Encoding

Performs the same encoding as the Pandas **factorize** method.

```python
from sklearn.preprocessing import OrdinalEncoder

protocol_type = X_train[['protocol_type']]

ordinal_encoder = OrdinalEncoder()
protocol_type_encoded = ordinal_encoder.fit_transform(protocol_type)
```

```python
# We show on the screen how they have been coded
for i in range(10):
    print(protocol_type["protocol_type"].iloc[i], "=", protocol_type_encoded[i])
```

```python
print(ordinal_encoder.categories_)
```

The problem with this type of encoding is that certain ML algorithms that work by measuring the similarity of two points by distance, will consider that 1 is closer to 2 than to 3, and in this case (for these categorical values), it does not make sense. Therefore, other categorisation methods are used, such as, for example, One-Hot encoding.

#### One-Hot Encoding

Generates for each category of the categorical attribute a binary array representing the value.

```python
# The sparse matrix only stores the position of values that are not '0' to save memory.
from sklearn.preprocessing import OneHotEncoder

protocol_type = X_train[['protocol_type']]

oh_encoder = OneHotEncoder()
protocol_type_oh = oh_encoder.fit_transform(protocol_type)
protocol_type_oh
```

```python
# Convert the sparse matrix to a Numpy array
protocol_type_oh.toarray()
```

```python
# We show on the screen how they have been coded
for i in range(10):
    print(protocol_type["protocol_type"].iloc[i], "=", protocol_type_oh.toarray()[i])
```

```python
print(ordinal_encoder.categories_)
```

In many occasions when partitioning the dataset or performing a prediction with new examples new values for certain categories appear that will cause an error in the **transform()** function. The OneHotEncoding class provides the **handle_uknown** parameter to either generate an error or ignore if an unknown categorical feature is present during the transformation (the default is to throw an error).

When this parameter is set to "ignore" and an unknown category is encountered during the transformation, the resulting coded columns for this feature will be all zeros. In the reverse transformation, an unknown category shall be denoted as None.




```python
oh_encoder = OneHotEncoder(handle_unknown='ignore')
```

## 5. Scaling the dataset

Before we start, we are going to retrieve the clean dataset and separate the labels from the rest of the data, we do not necessarily want to apply the same transformations on both sets.

```python
X_train = train_set.drop("class", axis=1)
y_train = train_set["class"].copy()
```

As a general rule, Machine Learning algorithms do not behave properly if the values of the features they receive as input are in very different ranges. For this reason, different scaling techniques are used. **It is important to note that these scaling mechanisms should not be applied on the labels.
* **Normalisation:** Attribute values are scaled to acquire a value between 0 and 1
* **Standardisation:** Attribute values are scaled and receive a similar value but do not fall within a range

**It is important to test these values by performing the transformations on the training data set only. Afterwards, they shall be applied on the test data set to test.**

```python
from sklearn.preprocessing import RobustScaler

scale_attrs = X_train[['src_bytes', 'dst_bytes']]

robust_scaler = RobustScaler()
X_train_scaled = robust_scaler.fit_transform(scale_attrs)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=['src_bytes', 'dst_bytes'])
```
